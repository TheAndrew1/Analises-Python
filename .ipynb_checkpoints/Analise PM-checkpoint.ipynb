```python
# Substituir a vírgula por ponto
df2022['happiness_score'] = df2022['happiness_score'].str.replace(',', '.', regex=False)

# Agora você pode converter a coluna para tipo numérico
df2022['happiness_score'] = pd.to_numeric(df2022['happiness_score'], errors='coerce')
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# Caminho local para o shapefile baixado
shapefile_path = 'paises/ne_110m_admin_0_countries.shp'

# Carregar o mapa mundial
world = gpd.read_file(shapefile_path)

# Garantir que os nomes dos países estejam alinhados
correcoes = {
    'United States': 'United States of America',
    'Czech Republic': 'Czechia',
    'Taiwan Province of China': 'Taiwan',
    'Bosnia and Herzegovina': 'Bosnia and Herz.',
    'Dominican Republic': 'Dominican Rep.',
    'Eswatini, Kingdom of': 'eSwatini',
    'North Cyprus': 'N. Cyprus',
    'Ivory Coast': "Côte d'Ivoire",
    'Palestinian Territories': 'Palestine',
    'Ivory Coast': "Côte d'Ivoire",
    # Adicione outras correções necessárias aqui
}
df2022['country'] = df2022['country'].replace(correcoes)

# Mesclar o dataframe `df2022` com o GeoDataFrame `world`
merged = world.merge(df2022, left_on='NAME', right_on='country', how='left')

# Categorizar os valores do happiness_score
# Usar qcut para dividir em 5 categorias baseadas em percentis
categories = pd.qcut(merged['happiness_score'], 5, labels=['Muito Baixo', 'Baixo', 'Médio', 'Alto', 'Muito Alto'])

# Atribuir essas categorias ao GeoDataFrame
merged['happiness_category'] = categories

# Criar uma versão ajustada da paleta 'Greens'
original_greens = plt.colormaps['Greens']
custom_greens = LinearSegmentedColormap.from_list(
    'CustomGreens', 
    original_greens(np.linspace(0.2, 1, 256))  # Ajuste o início para 0.2 (ou outro valor)
)

# Criar o mapa temático baseado nas categorias
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged.plot(
    column='happiness_category',  # Coluna com as categorias
    cmap='RdYlGn',               # Paleta de cores
    #cmap=custom_greens,
    legend=True,                  # Exibir legenda
    legend_kwds={'title': 'Níveis de Felicidade'},
    ax=ax,
    missing_kwds={'color': 'lightgrey', 'label': 'Sem dados'},
    edgecolor='dimgray',
    linewidth=0.3
)

# Personalizar o gráfico
ax.set_title('Distribuição Global da Felicidade em 2022: Uma Visão por Níveis', fontsize=16)
ax.set_axis_off()  # Remove os eixos
plt.show()
```

```python
# Salvar o gráfico como uma imagem
fig.savefig('felicidade_mundial_2022.png', dpi=300, bbox_inches='tight', format='png')
```

```python
# Inicializar uma lista para armazenar os dataframes
df_list = []

# Lista de anos (2015 a 2022)
years = range(2015, 2023)

# Para cada ano, selecionar os 10 primeiros países pela coluna 'happiness_rank' e adicionar a coluna 'year'
for year in years:
    df = globals()[f'df{year}']  # Acessar o dataframe correspondente ao ano
    top_10 = df.nsmallest(10, 'happiness_rank')  # Selecionar os 10 primeiros países pela coluna 'happiness_rank'
    top_10['year'] = year  # Adicionar a coluna 'year'
    df_list.append(top_10[['happiness_rank', 'country', 'year']])  # Selecionar as colunas desejadas

# Concatenar todos os dataframes na lista em um único dataframe
final_df = pd.concat(df_list, ignore_index=True)

# Exibir o dataframe final
print(final_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supondo que final_df já tenha sido criado corretamente
# Agrupar os dados para ter uma linha para cada país com suas posições ao longo dos anos
df_pivot = final_df.pivot(index='year', columns='country', values='happiness_rank')

# Dicionário de mapeamento para os países em português
country_translation = {
    'Australia': 'Austrália',
    'Austria': 'Áustria',
    'Canada': 'Canadá',
    'Denmark': 'Dinamarca',
    'Finland': 'Finlândia',
    'Iceland': 'Islândia',
    'Luxembourg': 'Luxemburgo',
    'Netherlands': 'Países Baixos',
    'New Zealand': 'Nova Zelândia',
    'Norway': 'Noruega',
    'Sweden': 'Suécia',
    'Switzerland': 'Suíça',
    # Adicione mais países conforme necessário
}

# Gerar uma paleta de cores com o número de países
palette = sns.color_palette("tab20", len(df_pivot.columns))

# Plotar o gráfico de linhas
plt.figure(figsize=(12, 8))

# Plotar as linhas sólidas
for idx, country in enumerate(df_pivot.columns):
    # Usar o nome traduzido para o país na legenda
    translated_country = country_translation.get(country, country)  # Caso não tenha tradução, usa o nome original

    sns.lineplot(data=df_pivot[country], color=palette[idx], lw=12, label=translated_country)

    # Plotar os pontos para valores únicos (aparecem uma única vez)
    years_with_data = df_pivot[country].dropna().index
    plt.plot(years_with_data, df_pivot[country].loc[years_with_data], 
             'o', markersize=15, color=palette[idx])  # Pontos com a mesma cor da linha

# Ajustar o gráfico
plt.title('Evolução do Ranking de Felicidade ao Longo dos Anos (2015-2022)', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Posição no Ranking de Felicidade', fontsize=12)
plt.xticks(df_pivot.index, rotation=45)  # Os anos estão agora no eixo X
plt.yticks(range(1, 11))  # Eixo Y de 1 a 10
plt.gca().invert_yaxis()  # Inverter o eixo Y para que 1 esteja no topo
plt.legend(title='Países', bbox_to_anchor=(1.05, 1), loc='upper left')  # Legenda ao lado

# Salvar o gráfico antes de exibi-lo
plt.tight_layout()
plt.savefig('ranking_felicidade_2015_2022.png', dpi=300)  # Salva a imagem em alta resolução

# Exibir o gráfico
plt.show()
```
